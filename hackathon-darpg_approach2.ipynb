```python
# Load the data from the JSON files
with open('/kaggle/input/hackathon-darpg/problem_statement_1_and_2/no_pii_grievance_v2/no_pii_grievance_v2.json', 'r') as data_file:
    data = json.load(data_file,parse_constant=None)
```

```python
import pandas as pd
import numpy as np

# Initialize an empty list to store rows
x = []

# Open the CSV file for reading
with open('/kaggle/input/mapping-darpg/Mapping.csv', 'r') as file:
    # Iterate over each row in the CSV file
    attr = file.readline().strip().split(';')
#     print(attr)
    for line in file:
        if line.isspace():
            continue
#         print(line)
        fields = line.strip().split(';')
        fields = [value if value != '' else np.nan for value in fields]
        row_data = {attr[i]: value for i, value in enumerate(fields)}
        x.append(row_data)
```

```python
df = pd.DataFrame(x)
df.head()
```

```python
df = df.dropna(subset=['OrgCode'])
```

```python
org_codes = df["OrgCode"].unique()
# invalid = ['16708','6357','6331','16706','6331','5621']

# org_codes = [x for x in org_codes if x not in invalid]
org_codes
```

```python
org_labels = {}
for i in range(len(org_codes)):
    org_labels[org_codes[i]] = i+1
    
org_labels['MORLY']
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
import spacy

# Load the English model
# nlp = spacy.load("en_core_web_sm")

nltk.download('stopwords')

delimiters = ['\r','\n','\t','-','?','>>', '/',',','.','!','&',';',':','(',')']

def preprocess(s):
    s = s.lower()
    for x in delimiters:
        s = s.replace(x," ")
    
    return s
```

```python
data[2]
```

```python
dataset = []
c=0
dummy_data = data[0:10]
for item in data:
    code = item['_id'][0:5]
    if 'subject_content_text' in item:
        if code in org_codes:
            mp = {}
            mp['text'] = preprocess(item['subject_content_text'])
            mp['label'] = org_labels[code]
            dataset.append(mp)
```

```python
dataset[0]
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
import pandas as pd

train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)

train_data
```

```python
train_data['label'] = train_data['label'].astype(int)
test_data['label'] = test_data['label'].astype(int)
```

```python
train_dict={'text': [x for x in train_data['text']],'label':[x for x in train_data['label']]}
```

```python
test_dict={'text': [x for x in test_data['text']],'label':[x for x in test_data['label']]}
```

```python
from transformers import AutoTokenizer
from datasets import DatasetDict, Dataset

test_dataset = Dataset.from_dict(test_dict)
train_dataset = Dataset.from_dict(train_dict)

dataset_dict = DatasetDict({'train': train_dataset, 'test': test_dataset})
```

```python
dataset_dict
```

```python
dataset_dict["train"][100]
```

```python
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/indic-bert")

def tokenize_function(examples):
    return tokenizer(examples["text"],padding="max_length", truncation=True,max_length=512)

tokenized_dataset = dataset_dict.map(tokenize_function, batched=True)
```

```python
tokenized_dataset
```

```python
# Load model directly
num_labels = len(org_codes)
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("ai4bharat/indic-bert",num_labels = num_labels )
```

```python
small_train_dataset = tokenized_dataset["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_dataset["test"].shuffle(seed=42).select(range(1000))
```

```python
small_eval_dataset
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch", num_train_epochs = 5,report_to=[])
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
